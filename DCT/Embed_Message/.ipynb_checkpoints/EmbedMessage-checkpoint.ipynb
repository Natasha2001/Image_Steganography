Step1: Import libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.fftpack import fft,dct
from itertools import product
from PIL import Image
```

Step2: Save the image in which we want to hide message in the same folder with image name as hide_image

```python
hide_path = !ls cover_image*
cover_img=cv2.imread(hide_path[0])
plt.imshow(cover_img)
```

Step3: DCT method of stegnography works on 8*8 pixel array so we need to ensure that our image pixels(height and width) are divisible by 8. If our image pixels are not divisible by 8 we resize the image

```python
height,width,channel=cover_img.shape
if height%8!=0:
    height=height+(8-(height%8))
if width%8!=0:
    width=width+(8-(width%8))
img=cv2.resize(cover_img,(width,height))
```

Step4: Split the image into RGB values

```python
bimg=img[:,:,0]
gimg=img[:,:,1]
rimg=img[:,:,2]
#display the images
output_grid = np.concatenate([np.array(bimg),np.array(gimg),np.array(rimg)], axis=1)
image_grid = Image.fromarray(output_grid)
image_grid
```

Step5: Save the secret message in message.txt file in the same folder

```python
f = open("message.txt", "r")#open the file and read the message
secret_message=f.read()
print(secret_message)
```

```python
secret=str(len(secret_message))+" "+secret_message
print(secret)
```

Step6: Convert each character in message to binary format and store them in a list 

```python
bits_list=[]
for char in secret:
    binary=bin(ord(char))#convert to binary
    bin_do=binary[2:]#remove first 2 charcaters from binary
    binval=bin_do.rjust(8,'0')#add sign bit
    bits_list.append(binval)
print(bits_list)
```

```python
bitmess=[]
for i in bits_list:
    list1=[]
    for j in i:
        list1.append(j)
    bitmess.append(list1)
```

Step7: Define the standard qunatization matrix

```python
quant = np.array([[16,11,10,16,24,40,51,61],
                 [12,12,14,19,26,58,60,55],
                 [14,13,16,24,40,57,69,56],
                 [14,17,22,29,51,87,80,62],                    
                 [18,22,37,56,68,109,103,77],
                 [24,35,55,64,81,104,113,92],
                 [49,64,78,87,103,121,120,101],
                 [72,92,95,98,112,100,103,99]])
```

Step8: Define function to convert to binary

```python
def binary_value(val):
    binval = bin(val)[2:]
    while len(binval) < 8:
        binval = "0"+binval
    return binval
```

Create a blank image to store the resultant blue image

```python
fin_img=np.zeros((bimg.shape),np.uint8)
```

Initialise counters

```python
messindex=0
letterindex=0
```

Step8: Hide the message in the blue part of the image using DCT method

```python
bimg=np.float32(bimg)
quantisedct=[]                 

x_blocks=width//8
y_blocks=height//8
if (x_blocks*y_blocks)<len(secret_message):
    print("Message too long cannot be encrypted")
for x in range(0,x_blocks):
    for y in range(0,y_blocks):
        x_offset=x*8
        y_offset=y*8
        x_end=x_offset+8
        y_end=y_offset+8
        imgblocks=bimg[y_offset:y_end,x_offset:x_end]-128#break the cover image into 88 block of pixels and subtract 128
        quantisedct=(cv2.dct(imgblocks))#apply DCT to each block 
        for i in range(0,8):
            for j in range(0,8):
                    quantisedct[i][j]=(np.round(quantisedct[i][j]/quant[i][j] ))#compress each block through quantization table 
        quantisedblock=quantisedct.copy()
        dc=quantisedblock[0][0]#find the value of the first pixel and convert it to binary
        dc_list=[]
        n=(int(dc))
        if n<0:
            n=abs(n)
            bin1=binary_value(n)
            dc="1"+bin1[1:8]
        if n>=0:
            dc=bin(n)[2:].rjust(8,'0')
        dc=str(dc)
        for i in dc:
            dc_list.append(i)
        if messindex<len(secret_message) and letterindex<8:#if some part of message is still left to be hidden
            dc_list[7]=bitmess[messindex][letterindex]#Replace the LSB of the first pixel with each bit of secret message. 
            letterindex+=1
            if letterindex==8:
                letterindex=0
                messindex+=1         
            makeitastring = ''.join(map(str, dc_list))
            dc=int(makeitastring,2)
            dc=np.float32(dc)
            quantisedblock[0][0]=dc#change the first pixel of the quantized block
        for i in range(0,8):
            for j in range(0,8):
                quantisedblock[i][j]=np.round(quantisedblock[i][j]*quant[i][j])#expand each block through quantization table
        simgblocks=cv2.idct(quantisedblock)#apply inverse DCT to each block
        simgblocks=simgblocks+128#add 128 to each block
        newblock=np.round(simgblocks)

        fin_img[y_offset:y_end,x_offset:x_end]=newblock
```

Step10: Create a blank image and copy the new blue part of image along with the red and green part in it

```python
nfin_img=np.zeros((img.shape),np.uint8)
nfin_img[:,:,0]=fin_img
nfin_img[:,:,1]=gimg
nfin_img[:,:,2]=rimg

plt.imshow(nfin_img)
```

Step11: Save the image

```python
cv2.imwrite("stego_img.png",nfin_img)
```
